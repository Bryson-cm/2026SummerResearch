```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 1000)

width = np.sqrt(1 - x**2)

z_upper = x + width
z_lower = x - width

fig, ax = plt.subplots(figsize=(7,7))

ax.fill_betweenx(
    x,
    z_lower,
    z_upper,
    alpha=0.5,
    color='cornflowerblue'
)

# beam trajectory in spacetime projection
ax.plot(x, x, 'k-', lw=3)

ax.set_xlabel('z')
ax.set_ylabel('x')
ax.set_aspect('equal')
plt.show()
```
