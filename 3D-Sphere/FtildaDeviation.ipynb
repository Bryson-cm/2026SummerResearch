```python
#Bryson Mariano
#Focal Length Theoretical Path
#Week 2-5
#Universal Parameters

import numpy as np
import math
import matplotlib.pyplot as plt




me    = 9.1093837 * 10 ** (-31)  #kg
c     = 299792458.0              #m/s
ne    = 1 * 10.0**15 *10**6                 #number density
e0    = 8.85418782 * 10**(-12)   #m^-3 kg^-1s^4 A^2
e     = 1.60217663 * 10**(-19)   #coulombs
wp    = ((ne*e**2)/(me*e0)) ** (1/2)


#Define Parameters


gamma = 110                      #Unitless
k     = 0.475 *me*wp**2          #kg/s^2
rb    = 0.65  *c/wp               #c/omega_p   
px    = 110   *me*c                #mec


f0=((px)**2) / (gamma * me * k)

f02=((px**2)*c/(k*(gamma*c*me+px)))


#Spherical Parameters
Rb    = 0.65 *c/wp

#Intialize y0, dy and x ###########################################

y0 = np.linspace(0, rb-(0.01*c/wp), 100)

################################################################

#Create/define Focal Length Function

def f_cyl(i):
    return (f0/(2*rb)) * (1 - (i / rb)**(2))**(-1/2)

def f_sph_old(i, j):
    return (f0/(Rb)) * (1 + (j/Rb)**2 - 2* (i / rb)**(2))**(-1/2)

def f_sph_new(i, j):
    return (f02/(Rb)) * (1 + (j/Rb)**2 - 2* (i / rb)**(2))**(-1/2)

def f_cyl_new(i):
    return (f02/(2*rb)) * (1 - (i / rb)**(2))**(-1/2)



#Create a Range of dy and y0

f1 = f_cyl(y0)
f2 = f_sph_old(y0, rb)
f3 = f_sph_new(y0, rb)
f4 = f_cyl_new(y0)

print(f3)
#Make a plot showing focal length vs different y0 values
print(f4[2]/f3[2])

plt.plot(y0/rb, f1*1000, color='black', label='Old Cylindrical Model')
plt.plot(y0/rb, f2*1000, color='blue', label='Old Spherical Model')
plt.plot(y0/rb, f3*1000, color='red', linestyle='--', label='New Spherical Model')
plt.plot(y0/rb, f4*1000, color='green', linestyle='--', label='New Cylindrical Model')
plt.title("Focal lentgh vs $y_0/r_b$ ratio for the cylindrical model")
plt.xlim(0, 1.0)
plt.ylim(10, 200)
plt.xlabel("$y_0/r_b$")
plt.ylabel("f (mm)")
plt.legend()
plt.show()




```
